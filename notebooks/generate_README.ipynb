```python
import pandas as pd

# 效褌械薪懈械 写邪薪薪褘褏
df = pd.read_csv('../data/diabetes.csv')

# README 
content = f"""# 袩褉芯械泻褌: 袗薪邪谢懈蟹 写懈邪斜械褌邪
袨锌懈褋邪薪懈械: 袠褋褋谢械写芯胁邪薪懈械 锌褉懈蟹薪邪泻芯胁 锌邪褑懈械薪褌芯胁 写谢褟 泻谢邪褋褋懈褎懈泻邪褑懈懈 蟹邪斜芯谢械胁邪薪懈泄.

## 袨锌懈褋邪薪懈械 写邪薪薪褘褏
- 袣芯谢懈褔械褋褌胁芯 蟹邪锌懈褋械泄: {df.shape[0]}
- 袣芯谢懈褔械褋褌胁芯 锌褉懈蟹薪邪泻芯胁: {df.shape[1]}

| 袩褉懈蟹薪邪泻 | 孝懈锌 | 袩褉芯锌褍褋泻懈 | 袦懈薪懈屑褍屑 | 袦邪泻褋懈屑褍屑 |
| :--- | :--- | :--- | :--- | :--- |
"""

# 袛芯斜邪胁谢械薪懈械 褋褌褉芯泻 褌邪斜谢懈褑褘 褑懈泻谢芯屑
for col in df.columns:
    content += f"| {col} | {df[col].dtype} | {df[col].isnull().sum()} | {df[col].min()} | {df[col].max()} |\n"

# 小芯褏褉邪薪械薪懈械 胁 泻芯褉械薪褜 锌褉芯械泻褌邪
with open('../README.md', 'w', encoding='utf-8') as f:
    f.write(content)

print("README.md 芯斜薪芯胁谢械薪!")
```
