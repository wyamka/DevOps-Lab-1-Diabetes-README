```python
import pandas as pd
import os

# Путь к файлу
file_path = '../data/diabetes.csv'

# Проверка наличия файла
if os.path.exists(file_path):
    # Чтение данных
    df = pd.read_csv(file_path)

    # README 
    content = f"""# Проект: Анализ диабета
Описание: Исследование признаков пациентов для классификации заболеваний.

## Описание данных
- Количество записей: {df.shape[0]}
- Количество признаков: {df.shape[1]}

| Признак | Тип | Пропуски | Минимум | Максимум |
| :--- | :--- | :--- | :--- | :--- |
"""

    # Добавление строк таблицы циклом
    for col in df.columns:
        content += f"| {col} | {df[col].dtype} | {df[col].isnull().sum()} | {df[col].min()} | {df[col].max()} |\n"

    # Сохранение в корень проекта
    with open('../README.md', 'w', encoding='utf-8') as f:
        f.write(content)

    print("README.md успешно обновлен!")
else:
    print(f"Ошибка: Файл по адресу {file_path} не найден. Проверьте папку data.")
```
